### Likes-toviews Bar Graph

```python
import plotly.graph_objects as go

bar_graphs = []
num_companies = len(companies)
plot_dicts = []
j = 0
initial_visible = [True] + [False] * (num_companies - 1)  # Only the first company is initially visible

for company, content in companies.items():
    company_likes = {}
    company_views = {}
    for pers in persuasions:
        company_likes[pers] = 0
        company_views[pers] = 0
    
    for video, values in content["videos"].items():
        strats = values["annots"]
        likes = values["likeCount"]
        views = values["viewCount"]
        
        for strat in strats:
            company_likes[strat] += likes
            company_views[strat] += views
    
    strat_names = list(company_likes.keys())
    like_to_views_ratios = []
    
    # Calculate the like-to-views ratio for each strategy, handling division by zero
    for strat in strat_names:
        if company_views[strat] == 0:
            ratio = 0  # Avoid division by zero
        else:
            ratio = company_likes[strat] / company_views[strat]
        like_to_views_ratios.append(ratio)
    
    # Normalize the like-to-views ratio using percentage share within the company
    total_ratio = sum(like_to_views_ratios)
    if total_ratio != 0:  # Avoid division by zero
        normalized_ratios = [ratio / total_ratio * 100 for ratio in like_to_views_ratios]
    else:
        normalized_ratios = [0] * len(strat_names)
    
    visible_action = initial_visible[j]  # Determine initial visibility based on 'initial_visible' list
    
    # Create a bar graph for the current company with normalized ratios
    bar_trace = go.Bar(
        x=strat_names,
        y=normalized_ratios,
        name=company,
        visible=visible_action,
    )

    bar_layout = go.Layout(
        title=f"Normalized Like-to-Views Ratio by Persuasion Strategy for {company}",
        xaxis_title="Persuasion Strategy",
        yaxis_title="Normalized Like-to-Views Ratio (%)",
    )
    
    bar_graphs.append((bar_trace, bar_layout))

    visibility = [company == company_name for company_name in companies.keys()]  # Set visibility based on the current company
    
    d = dict(args=[{"visible": visibility}, {"title": company + " (" + str(content["count"]) + ")"}], 
             label=company + " (" + str(content["count"]) + ")", method="update")
    plot_dicts.append(d)
    
    j += 1

fig = go.Figure(data=[bar_trace for bar_trace, _ in bar_graphs], layout=bar_layout)
fig.update_layout(
    updatemenus=[dict(buttons=plot_dicts,
                      pad={"r": 10, "t": 10},
                      showactive=True,
                      x=1,
                      xanchor="left",
                      y=1.2,
                      yanchor="top")]
)

fig.show()
```
